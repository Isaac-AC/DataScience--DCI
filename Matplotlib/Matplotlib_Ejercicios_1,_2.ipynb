# Ejercicio 1

Cre un conjunto de graficas como la que se muestra en la imagen y en cada una utiliza puntos con disperci√≥n

Imagen: image.png

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 15, 1000)
y = x + np.random.randn(1000)
```

```python
plt.plot(x,y, '.')
plt.xlim(0,15)
plt.ylim(0,15)
```

```python
data = x,y
```

```python
fig = plt.figure(constrained_layout=True)
gs = fig.add_gridspec(3, 3)
ax1 = fig.add_subplot(gs[0, :])

ax1.set_title('gs[0, :]')
ax1.plot(x,y,'.',c='b')

ax2 = fig.add_subplot(gs[1, :2])
ax2.set_title('gs[1, :2]')
ax2.plot(x, y,'.',c='r')

ax3 = fig.add_subplot(gs[1:, 2])
ax3.set_title('gs[1:, 3]')
ax3.plot(x**2,y,'.',c ='pink')

ax4 = fig.add_subplot(gs[2, 0])
ax4.set_title('gs[2, 0]')
ax4.plot(x,y**2,'.',c='y')

ax5 = fig.add_subplot(gs[2, 1])
ax5.set_title('gs[2, 1]')
ax5.plot(x,y,'.', c='g')
ax5.set_xscale('log')

#fig.show()
```

# Ejercicio 2

Cambiar todos los pixeles que esten en negro de una imagen a algun color

```python
from PIL import Image
```

```python
imgURL = "https://github.com/alrakomala/DataScienceDCI/blob/main/Matplotlib/Full_Moon_Luc_Viatour.jpg?raw=true"
```

```python
import requests
im = Image.open(requests.get(imgURL, stream=True).raw)
```

```python
arrayIm = np.array(im)
arrayIm.shape
```

```python
for i in range(arrayIm.shape[0]):
  for j in range(arrayIm.shape[1]):
    if (arrayIm[i,j] == [0,0,0]).any():
      arrayIm[i,j] = [0,255,0]     
```

```python
img = Image.fromarray(arrayIm, 'RGB')
img
```
