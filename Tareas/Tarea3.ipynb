```python
# Using API REST

endpoint = "https://es.wikipedia.org/w/rest.php/v1/search/page?summary"
params = {
    "q": "teoría",
    'limit': 20
        }

r1 = requests.get(endpoint, params=params)
r1.status_code
```

```python
r1.json()
```

```python
df = pd.json_normalize(r1.json()['pages'])
```

```python
# Elimina None
df["description"] = df["description"].fillna("0")
```

```python
df2 = pd.DataFrame()
word = list()
```

```python
l = a = 0
for i in df["description"]:
    if i.find("física") != -1:
        a = 1
        word.append("física")
    elif i.find("ciencia") != -1:
        a = 1
        word.append("ciencia")
    if a == 1:
        df2 = df2.append(df.iloc[[l]], ignore_index=True)
    a = 0
    l = 1 + l
```

```python
df2["Word"] = word
df2 = df2[["id", "title", "description", "Word"]]
df2
```
