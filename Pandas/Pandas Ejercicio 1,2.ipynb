```python
import matplotlib.pyplot as plt
```

```python
#!pip install opendatasets
#import opendatasets as od
#od.download("https://www.kaggle.com/datasets/thedevastator/global-fossil-co2-emissions-by-country-2002-2022")
```

```python
data = pd.read_csv("GCB2022v27_MtCO2_flat.csv")
data.head(5)
```

```python
data.columns
```

```python
dataMx = data.loc[data["Country"] == "Mexico"]
dataMx.head(5)
```

```python
dataMx = dataMx[:].fillna(0)
```

```python
dataMx.head()
```

```python
dataMx.describe()
```

```python
plt.bar(dataMx['Year'], dataMx['Coal'])
plt.ylabel('Coal emission')
plt.xlabel('Year')
```

```python
dataMxreal = data.loc[(data["Country"] == "Mexico") & (data["Year"] > 1890)]
dataMxreal = dataMxreal[:].fillna(0)
dataMxreal.head(5)
```

```python
plt.bar(dataMxreal['Year'], dataMxreal['Total'])
plt.ylabel('Total')
plt.xlabel('Year')
```

```python
dataMxreal = dataMxreal.set_index("Year")
```

```python
Sections = dataMxreal.columns[2:]
Sections = Sections.values.tolist()
Sections
```

```python
# Define the number of rows and columns for the subplots
nrows = 4
ncols = 2

# Create a new figure and axis objects
fig, axs = plt.subplots(nrows=nrows, ncols=ncols, figsize=(12, 18))

# Flatten the axs array so we can iterate over it with a single index
axs = axs.flatten()

# Use a for loop to create a subplot for each column in the dataframe
for i, column in enumerate(Sections):
    axs[i].plot(dataMxreal[column])
    axs[i].set_title(column)

# Set the tight layout to minimize overlapping of subplots
plt.tight_layout()

# Show the figure
plt.show()
```
